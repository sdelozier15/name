```python
import pandas as pd
import seaborn as sns
from pandas.plotting import scatter_matrix
from matplotlib import pyplot
```

```python
data = pd.read_csv("FINAL_ML.csv")
df = pd.DataFrame(data)
df2 = df.drop(columns = "Unnamed: 0")
df2.columns
```

```python
ml = df2[['Outcome', 'pedmort_2016', 'pedmort_53mean', 'pedmort_3mean', 'Code', 'sanit_2015',
       'sanit_21mean', 'sanit_3mean',
       'democ_2018', 'latitude', 'within_tropics']]
ml
```

```python
sns.heatmap(ml.corr(), annot=True, fmt=".2f")
plt.show()
```

```python
ax = sns.boxplot(data=ml, orient="h", palette="Set2")
```

```python
# histograms
ml.hist()
pyplot.show()
```

```python
# scatter plot matrix
scatter_matrix(ml)
pyplot.show()
```
