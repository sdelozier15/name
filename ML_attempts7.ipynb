```python
col = le_df.loc[: , "2016":"2018"]
le_df['le_3mean'] = col.mean(axis=1)
le_df.head()
```

```python
col2 = le_df.loc[: , "1950":"2015"]
le_df['le_65mean'] = col2.mean(axis=1)
le_df.rename(columns = {"2019": "le_2019"})
```

```python
le_final = le_df[['2019', 'le_3mean', 'le_65mean']]
le_final.head()
```

```python
le_final2 = le_final.rename(columns={'2019':'le_2019'})
le_final2
```

```python
code = le_df[['Code']]
code.head()
```

```python
col3 = le_df.loc[: , "1950":"2019"]
col_ranges = col3.max(axis=1) - col3.min(axis=1)
le_69roc = col_ranges / 69
roc_df1 = pd.concat([le_69roc, code], axis = 1)
roc_df = roc_df1.rename(columns={0:'le_69roc'})
roc_df 
```

```python
le_finals1 = pd.concat([le_final2, code], axis=1)
le_finals = le_finals1.merge(roc_df , on='Code', how='inner')
le_finals.head()
```
